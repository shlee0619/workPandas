```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 정의
solved_problems = np.array([17, 13, 12, 15, 16, 14, 16, 16, 18, 19])
attitude = np.array([94, 73, 59, 80, 93, 85, 65, 79, 77, 91])

# 피어슨 상관계수 계산
mean_x = np.mean(solved_problems)
mean_y = np.mean(attitude)

# 편차와 편차 곱
deviation_x = solved_problems - mean_x
deviation_y = attitude - mean_y
numerator = np.sum(deviation_x * deviation_y)

# 분모
denominator = np.sqrt(np.sum(deviation_x**2) * np.sum(deviation_y**2))

# 피어슨 상관계수
pearson_corr = numerator / denominator

# 산점도 그리기
plt.scatter(solved_problems, attitude)
plt.title("Scatter Plot of Solved Problems vs. Attitude")
plt.xlabel("Solved Problems")
plt.ylabel("Attitude")
plt.grid(True)
plt.show()

pearson_corr
```
